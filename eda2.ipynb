```python
data = pd.read_csv('df_merged.csv') # 4000개 데이터
```

```python
df2 = pd.read_csv('e약은요.csv') # 약 내용 데이터 
df2.drop(['제품명','공개일자','수정일자','업체명','이 약은 어떻게 보관해야 합니까?','이 약을 사용하기 전에 반드시 알아야 할 내용은 무엇입니가?'],axis=1,inplace=True)
```

```python
keys = ['이 약의 효능은 무엇입니까?','이 약은 어떻게 사용합니까?','이 약을 사용하기 전에 반드시 알아야 할 내용은 무엇입니가?','이 약의 사용상 주의사항은 무엇입니까?'
        ,'이 약을 사용하는 동안 주의해야 할 약 또는 음식은 무엇입니까?','이 약은 어떤 이상반응이 나타날 수 있습니까?']
values = ['효능','사용 방법','알아야 할 내용','주의사항','주의음식','부작용','성분명(영어)','제조사명','제조사명(영어)','약품 분류','품목기준 코드','전문의약품 여부','EDI코드','제형',
'모양','식별문자(앞)','식별문자(뒤)','색상','정제 분류명','영역']
my_dict = dict(zip(keys, values))
merge_df_4000 = merge_df_4000.rename(columns=my_dict)
merge_df_4000.notnull().sum().tail(7)
```

### 병용 금지,주의 약품 data

```python
df_stop= pd.read_csv('병용금지약물(2022).csv',encoding='cp949')
df_stop.drop_duplicates(subset='제품코드A', inplace=True)
df_stop.drop_duplicates(subset='제품코드B', inplace=True)
df_stop['제품코드A'] = df_stop['제품코드A'].astype(str)
df_stop['제품코드B'] = df_stop['제품코드B'].astype(str)
df_stop = df_stop[['제품코드A','제품코드B','상세정보']]
df_stop
```

```python
df_caution = pd.read_csv('병용주의약물(2021).csv',encoding='cp949')
df_caution = df_caution[['의약품코드A','의약품코드B','비고']]
df_caution.drop_duplicates(subset='의약품코드A', inplace=True)
df_caution.drop_duplicates(subset='의약품코드B', inplace=True)
df_caution['의약품코드A'] = df_caution['의약품코드A'].astype(str)
df_caution['의약품코드A'] = df_caution['의약품코드A'].astype(str)
df_caution.rename(columns = {'의약품코드A' : '제품코드A' , '의약품코드B' : '제품코드B' , '비고':'상세정보'}, inplace = True)
df_caution
```

```python
df_concat_warming =pd.concat([df_stop, df_caution], keys=['stop', 'caution'],ignore_index=True)
df_concat_warming
```

```python
# 문제될수있는 성분을 포함하는 이미지를 추출해줌 
index_1 = merge_df_4000[merge_df_4000['di_edi_code'].isin(df_concat_warming['제품코드A'].values)]['di_edi_code'].index # 문제될수 있는 성분1이 있는 index
index_2 = merge_df_4000[merge_df_4000['di_edi_code'].isin(df_concat_warming['제품코드B'].values)]['di_edi_code'].index # 문제될수 있는 성분2이 있는 index 
# merge_df_4000['New Variable'] = merge_df_4000.index.isin(index_values).map(lambda x: True if x else False)

merge_df_4000['문제가능'] = None
merge_df_4000.loc[index_1,'문제가능'] = 'problem'
merge_df_4000.loc[index_2,'문제가능'] = 'problem'
merge_df_4000['문제가능']
```
