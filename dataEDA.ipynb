## 구분자(,처리)

```python
# ,이 컬럼내 ,가 포함되어있는 행 몇개씩인지 추출해줌
def count_commas(df):
    count_dict = {}
    for column in df.columns:
        count = df[column].astype(str).str.count(',')
        count_dict[column] = count.sum()   # dict value값 -> 개수 
    return count_dict

count_commas(df)

#
```

```python
# str이랑 섞여있는것들 str타입으로 변경 
def convert_numeric_to_string(df, column):
    df[column] = df[column].astype(str)
    return df
convert_numeric_to_string(df,'EDI코드')
```

```python
# 각각 ','포함하는것들이 어떻게 생겼는지 확인 
df[df['dl_custom_shape'].astype(str).str.contains(',')]['dl_custom_shape'].unique()
```

```python
# ,를 공백으로 변경해야 할 컬럼들 적용
def replace_string(df, column, old_value, new_value):
    df[column] = df[column].str.replace(old_value, new_value)
    return df

replace_string(df,'제품명', ',' , ' ')
replace_string(df,'제품명(영어)', ',' , ' ')
replace_string(df,'식별문자(앞)', ',' , '')
```

```python
# ,를 |로 변경해야 할 컬럼들 적용
def replace_string(df, column, old_value, new_value):
    df[column] = df[column].str.replace(old_value, new_value)
    return df

replace_string(df,'dl_custom_shape', ',' , '|')
replace_string(df,'약품 분류', ',' , '|')
replace_string(df,'EDI코드', ',' , '|')
replace_string(df,'색상', ',' , ' ')
replace_string(df,'정제 분류명', ',' , ' ')
```

```python
# ,이 컬럼내 .가 포함되어있는 행 몇개씩인지 추출해줌
def count_commas(df):
    count_dict = {}
    for column in df.columns:
        count = df[column].astype(str).str.count('\.')
        count_dict[column] = count.sum()   # dict value값 -> 개수 
    return count_dict

count_commas(df)

#
```
