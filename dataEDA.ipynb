## 구분자(,처리)

```python
# 데이터프레임 전체 컬럼내에 str이 몇개씩 있는지 확인 
def count_commas(df,str):
    count_dict = {}
    for column in df.columns:
        count = df[column].astype(str).str.count(str)
        count_dict[column] = count.sum()   
    return count_dict


#
```

```python
# str이랑 섞여있는것들 str타입으로 변경 
def convert_numeric_to_string(df, column):
    df[column] = df[column].astype(str)
    return df
convert_numeric_to_string(df2,'di_edi_code')
```

```python
# 각각 ','포함하는것들이 어떻게 생겼는지 확인 
df2[df2['form_code_name'].astype(str).str.contains(',')]['form_code_name'].unique()
```

```python
# ,를 공백으로 변경해야 할 컬럼들 적용
def replace_string(df, column, old_value, new_value):
    df[column] = df[column].str.replace(old_value, new_value)
    return df

replace_string(df2,'dl_name', ',' , ' ')
replace_string(df2,'dl_name_en', ',' , ' ')
replace_string(df2,'print_front', ',' , '')
```

```python
# ,를 |로 변경해야 할 컬럼들 적용
def replace_string(df, column, old_value, new_value):
    df[column] = df[column].str.replace(old_value, new_value)
    return df

replace_string(df2,'dl_custom_shape', ',' , '|')
replace_string(df2,'di_class_no', ',' , '|')
replace_string(df2,'di_edi_code', ',' , '|')
replace_string(df2,'color_class1', ',' , '|')
replace_string(df2,'form_code_name', ',' , '|')
```

```python
# ,이 컬럼내 .가 포함되어있는 행 몇개씩인지 추출해줌
def count_commas(df):
    count_dict = {}
    for column in df.columns:
        count = df[column].astype(str).str.count('\.')
        count_dict[column] = count.sum()   # dict value값 -> 개수 
    return count_dict

count_commas(df2)

#
```

```python
# 한슬사이에.이 끼워져있으면 제거해줘라 
import re

def remove_periods_except_mixed_korean(df, column):
    pattern = r'(?<![가-힣])\.'
    df[column] = df[column].astype(str).apply(lambda x: re.sub(pattern, '', x))
    return df

remove_periods_except_mixed_korean(df2,'dl_material')
remove_periods_except_mixed_korean(df2,'dl_material_en')
remove_periods_except_mixed_korean(df2,'dl_name')
remove_periods_except_mixed_korean(df2,'dl_name_en')
```

```python
# 맞춤형 변경
def replace_string(df, column, old_value, new_value):
    df[column] = df[column].str.replace(old_value, new_value)
    return df.head()

replace_string(df2,'di_class_no', '.' , '|')
replace_string(df2,'dl_material', '.' , '|')
replace_string(df2,'dl_material_en', '.' , '|')
replace_string(df2,'dl_name', '.' , '|')
replace_string(df2,'dl_name_en', '.' , '|')
```

```python
df2['di_edi_code'] = df2['di_edi_code'].astype(str).replace(r'\.0$', '', regex=True) #공백제거
df2['di_edi_code'].value_counts().index
```

```python
# 공백제거
df2['leng_short'] = df2['leng_short'].replace(r'\s', '', regex=True)
df2['leng_long'] = df2['leng_long'].replace(r'\s', '', regex=True)
df2['thick'] = df2['thick'].replace(r'\s', '', regex=True)
```

```python
def separate_korean_nonkorean(df, column):
    df[column + '_korean'] = df[column].str.extract(r'([가-힣]+)', expand=False)
    df[column + '_nonkorean'] = df[column].str.extract(r'([^가-힣]+)', expand=False)
    print(df[[column + '_korean', column + '_nonkorean']])
    
separate_korean_nonkorean(df2, 'dl_name')
```
